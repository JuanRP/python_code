```python
# tower(base, height, modulus)

#If modulus == 1, the returned value should be 0 irrespective of the values of the two other parameters.
#Otherwise, if base == 1 or height == 0, the output should be 1, and, if height == 1, the output should be base % modulus.

#Algorithm:
#To compute (BP) % M,
#where base B and modulus M are positive integers,
#and P is a positive integer or a "power tower" of 2 or more terms each greater than 1:

#If M is 1, return 0
#If P is 1, return B % M
#Otherwise P is an integer greater than 1, or a power tower:
#    Calculate B0 % M, B1 % M, B2 % M, through BM % M,
#    then continue until you find the least N such that N>M and BN â‰¡ BM mod M.
#    Then C = N-M is the length of the cycle for powers of B mod M.
#    
#    If P is an integer or a power tower less than N+1, then BP % M is one of the values already examined in determining C.
#    If P is a power tower greater than N, invoke this algorithm recursively to get the value of P % C.
#    Then use P % C to determine which term in the series [BM % M, B(M+1) % M, ...] is equal to BP % M.
```
