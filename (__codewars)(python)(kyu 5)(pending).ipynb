```python
#tipo_agrupacion = cuadrilateros de 1 cuadro, cuadrilateros de 2 cuadrados, etc...
#perimetro = (numero de cuadrilateros por tipo_agrupacion) * (tipo_agrupacion * 2 + 2)
# 2 cuadrados  6 lados
# 3 cuadrados  8 lados
# 4 cuadrados 10 lados

##===============================
##===============================
##identifica las islas por coordenada
def islas (arr):
    isl = []
    islS = set()
    islD1 = {}
    indx = 0
    indy = 0
    cnt = 0
        
    for ar in arr:
        for ar1 in ar:
            if ar1 == 'X':
                isl.append ([indx,indy])
                islS.add(tuple([indx,indy]))
                islD1[tuple([indx,indy])] = cnt
            indy += 1
            cnt +=1
        indx += 1
        indy = 0
        cnt +=1
    return isl, islS, islD1
##===============================

##===============================
##===============================
##calcula distancia entre islas
def Cdist (p1, p2):
    d = ((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)**.5
    return d
##===============================




##===============================
##===============================
##== core code ==================
##===============================
##===============================
def land_perimeter(arr):
    
    l = islas (arr)
    lP = l[0]
    lS = l[1]
    lD1 = l[2]
    
    #=========================
    #=========================
    #Land Perimeter   lpm01
    #identificar puntos "que forman figuras 
    lS2 = set()
    pS = set()
    chkD = {}
    
    for lp1 in lP [0:len(lP)]:
        for lp2 in lP [1:len(lP)]:
            pd = Cdist (lp1, lp2)
            
            if pd == 1:
                # puntos NO aislados
                pS.add (tuple(lp1))
                pS.add (tuple(lp2))
                
                a1 = lD1[tuple(lp1)]
                a2 = lD1[tuple(lp2)]
                
                if tuple((tuple(lp1),tuple(lp2))) not in chkD or tuple((tuple(lp1),tuple(lp2))) not in chkD:
                    if lp1[0] < lp2[0]:
                        chkD [tuple((tuple(lp1),tuple(lp2)))] = 2
                        lS2.add(tuple((a1,a2)))
                        
                    elif lp1[0] == lp2[0] and lp1[1] < lp2[1]:
                        chkD [tuple((tuple(lp1),tuple(lp2)))] = 2
                        lS2.add(tuple((a1,a2)))
                    
                    else:   
                        chkD [tuple((tuple(lp2),tuple(lp1)))] = 2
                        lS2.add(tuple((a2,a1)))
                       
    lP2 = sorted(lS2)
    
    
    print ('lP2:',lP2)
    print ('=================')
    #lP2: [(1, 8), (8, 15), (14, 15), (15, 22), (22, 23), (23, 24), (23, 30), (44, 51), (57, 64), (67, 68)]
    
    lF = {}
    lrf = lP2[0]
    sm = 1
    ind = 0
    
    for lp in lP2 [1:len(lP2)]:
        print (lrf, lp, lp[1], lrf[0], lrf[1])
        if (lp[0] == lrf[0]) or (lp[0] == lrf[1]) or (lp[1] == lrf[1]):
            sm += 1
            print('pasa01')
        else:
            if sm not in chkD:
                #x = lF[sm]
                #print (x)
                lF[sm] =  1
                print('pasa02')
            else:
                lF[sm] = 1
                print('pasa03')
        lrf = lp
    
    print('lF:',lF)
    
    #Land Perimeter   lpm01
    lpm01 = 0
    for key in lF:
        lpm01 = key * 4 + lpm01
        
    #=========================
    
    #=========================
    #=========================
    #Land Perimeter   lpm01
    ##perimetro puntos aislados
    #identificar puntos "aislados"  //  puntos totales - puntos NO aislados 
    pSa = lS - pS
    NpSa = len (pSa)
    lpm02 = NpSa * 4
    #=========================
    
    out = lpm01 + lpm02
    return  ('Total land perimeter: ' + str(out))
```
